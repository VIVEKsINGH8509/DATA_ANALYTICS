```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate x-axis values ( 1 to 9)
x = np.arange(1, 10)
```

```python
# Generate y-axis values ( 1 to 18)
y = np.arange(1, 19)
```

```python
# select the first nine elements from y to match
y = y[:9]        # slicing to get elements from index
```

```python
# create thr line plot

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.scatter(x,y,s=50)             #increase marker size to 50
plt.show()
```

```python
#Exporting a plot
plt.plot(x,y)
plt.savefig('example.png')
```
