```python
# Group by functionality :
# 1. Splitting : the data is first split into groups based on the criteria provided
# 2.Applying : function is applied to each group inddependently
# 3.Combining : the result of the function is combined and returned into a new dataframe
```

```python
# Spliting first step
df_grouped = df.groupby('gender')
```

```python
#output will not have any visible impact except for
print(df_grouped)
```

```python
# 2. Applyibng
maean_tip_by_gender = df_grouped['tip'].mean()
print(maean_tip_by_gender)

print(df_grouped['tip'].min())
print(df_grouped['tip'].max())
print(df_grouped['tip'].std())
print(df_grouped['tip'].count())
print(df_grouped['tip'].median())
print(df_grouped['tip'].sum())
print(df_grouped['tip'].var())
```

```python
#  creating group on multiple columns
df_grouped = df.groupby(['gender', 'day'])
print(df_grouped['tip'].mean())
```

```python
# agg function
#Apply multiple agg functioin to different columns when grouping data
#Calculate both the sum and the index of the "tip" column for each combination of
#'Gender' & 'day'

# Group the df by both 'gender' & 'day' and calculate the mean and sum of 'tip' for
# each combination
df_grouped = df.groupby(['gender', 'day']).agg({'tip': ['mean', 'sum']}).reset_index()
print(df_grouped)
```

```python
# df_grouped = df.groupby(['gender', 'day']).agg({'tip': ['mean', 'sum']})
# print(df_grouped)
```

```python
#  Rename the column for clarity
df_grouped.columns = ['gender', 'day', 'mean_tip', 'sum_tip']
print(df_grouped)
```
