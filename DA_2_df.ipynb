```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/MS_Dhoni_ODI_record.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
#Data cleaning- Opposite name says 'v Aus' etc, we can remove 'v '
df['opposition'] = df['opposition'].str.replace('v ','', regex=False)
# df['Opposition'] = df['opposition'].apply(lambda x: x[2: ])
#regex=False means that the first string('v ') is not a regex, but a literal String
df
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst= True)
df['Year'] = df['date'].dt.year.astype(int)
```

```python
df
```

```python
df.Year  # or df['Year']
```

```python
# batting avg = total runs scored /(Innigs count - not outs count)

df['score'] = df['score'].apply(str)
df['not_out'] = np.where(df['score'].str.endswith('*'),1,0)
df
```

```python
#Removing odi_number column from the df.
# df.drop(columns='odi_number',inplace=True)
# df
```

```python
df_new = df.loc[((df['score']!='DNB') & (df['score'] != "TDNB")),'runs_scored':]
df_new
```

```python
# fixing the data types of numerical columns
df_new.info()
df_new['runs_scored'] =df_new['runs_scored'].astype(int)
df_new['balls_faced'] =df_new['balls_faced'].astype(int)
df_new['strike_rate'] =df_new['strike_rate'].astype(float)
df_new['fours'] =df_new['fours'].astype(int)
df_new['sixes'] =df_new['sixes'].astype(int)
```

```python
df_new.info()
```

```python
first_match_date = df['date'].dt.date.min().strftime('%B %d, %Y')
print(first_match_date)

last_match_date = df['date'].dt.date.max().strftime('%B %d, %Y')
print(last_match_date)
```

```python
numOfmatches = df.shape[0]
print(f"Number of matches : {numOfmatches}")

numOfinnings = df_new.shape[0]
print(f"Number of innings : {numOfinnings}")

not_outs = df_new['not_out'].sum()
print(f"Number of innings : {not_outs}")

runs_scored = df_new['runs_scored'].sum()
print(f"runs_scored : {runs_scored}")

balls_faced   = df_new['balls_faced'].sum()
print(f"balls_faced   : {balls_faced}")
```

```python
# batting avg = total runs scored /(Innigs count - not outs count)
career_sk = runs_scored / balls_faced
print("Career strike: {:.2f}".format(career_sk))

career_avg = runs_scored /(numOfinnings - not_outs )
print("Career avg: {:.2f}".format(career_avg))
```

```python
#no. of matches played against different opposition
#count the occurence of each unique value in the 'opposition' column
#opposition count will be a series with a labelled index as opposition
opposition_count = df['opposition'].value_counts()
opposition_count
```

```python
#Plot the counts as the bar plot
opposition_count.plot(kind='bar', title='No. of matches against diff oppositions', figsize=(8, 5))
plt.show()
```

```python
#Runs scored against each team
#group the df by opposition column
grouped_by_opposition = df_new.groupby('opposition')
#Sum the runs scored columns for each group
sum_of_runs_scored = grouped_by_opposition['runs_scored'].sum()
sum_of_runs_scored
```

```python
#sum of run scored is a series with a labelled index, which is opposition
#convert it into a df and remove the index
runs_scored_by_opposition = pd.DataFrame(sum_of_runs_scored).reset_index()
runs_scored_by_opposition.plot(kind='bar', x='opposition', y='runs_scored', title='Runs scored by opposition', figsize=(8, 5))
plt.show()
```
