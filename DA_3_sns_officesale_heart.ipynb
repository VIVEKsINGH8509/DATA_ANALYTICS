```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
# Create Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(x= x,y = y, color = 'blue', alpha = 0.6)
plt.title('Scatter Plot', fontsize = 15)
plt.xlabel('x' , fontsize = 12)
plt.ylabel('Y' , fontsize  = 12)

plt.grid(True)
plt.show()

```

```python
# cREATE sCATTER PLOT WITH sEABORN

sns.set_style('darkgrid')   #Set style with white grid background
plt.figure(figsize=(8,6))
sns.scatterplot(x=x,y=y)
plt.title('Scatter Plot', fontsize = 15)
plt.xlabel('x' , fontsize = 12)
plt.ylabel('y' , fontsize  = 12)



plt.show()
```

```python
# MAtplotlib
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html

fig , ax = plt.subplots()
ax.plot(ages['age'],ages['chol'] , ls = ":", lw = 1.7)
ax.set_xlabel('Age')
ax.set_ylabel('chol')
ax.set_title('Line Plot')
plt.show()
```

```python
df = pd.read_csv('/content/dm_office_sales.csv')
df
```

```python
#  Contplot : a simple plot , it merely shows the total count of rows per category
# a histogram
plt.figure(figsize=(10,4), dpi = 199)         # dpi  =dots per inch of space
sns.countplot(x = 'division' , data = df)
plt.show()


```

```python
plt.figure(figsize=(10,4),dpi = 200)
sns.countplot(x = 'level of education' , data = df)
plt.show()
```

```python
# Breaak down within anoher category with Hue
plt.figure(figsize=(10,4),dpi = 200)
sns.countplot(x = 'level of education' , data = df , hue = 'training level' )
plt.show()
```

```python
#using palette
plt.figure(figsize=(10,4),dpi = 200)
sns.countplot(x = 'level of education' , data = df , hue = 'training level' , palette = 'Set1')
plt.show()
```

```python
#using palette
plt.figure(figsize=(10,4),dpi = 200)
sns.countplot(x = 'level of education' , data = df , hue = 'training level' , palette = 'Set2')
plt.show()
```

```python
#using palette
plt.figure(figsize=(10,4),dpi = 200)
# pair would be a good choice if there was a distinct jump from 0 and 1 to 2 and 3
sns.countplot(x = 'level of education' , data = df , hue = 'training level' , palette = 'Paired')
plt.show()
```
