```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew
```

```python
#Generate random data with a larger sample size
sample_size = 100000
ratings = np.random.randint(1, 6, sample_size)
```

```python
#Modify data for right skew
right_skew_data = np.concatenate((ratings, np.random.randint(1, 2, sample_size//2)))
```

```python
#Modify data for left skew
left_skew_data = np.concatenate((ratings, np.random.randint(4, 6, sample_size//2)))
```

```python
#Function to calc mean, median, mode and skewness
def calculate_stats(data):
    mean = np.mean(data)
    median = np.median(data)
    skewness = skew(data)
    return mean, median, skewness
```

```python
#Function to plot mean and median with histogram
def plot_histogram(ax, data, title):
    ax.hist(data, bins=np.arange(1,7)-0.5, rwidth=0.8, alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel('Ratings')
    ax.set_ylabel('Frequency')
    ax.set_xticks(range(1,6))
    ax.grid(axis='y', alpha=0.5)

    #calc mean, median and skewness
    mean, median, skewness = calculate_stats(data)
    ax.axvline(x=mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')
    ax.axvline(x=median, color='blue', linestyle='--', label=f'Median: {median:.2f}')
    ax.legend()

    #Print mean, median and skewness
    print(f"{title}: ")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Skewness: {skewness:.2f}")
    print()

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
plot_histogram(axs[0], ratings, '')
plot_histogram(axs[1], right_skew_data, 'Right Skewed Data')
plot_histogram(axs[2], left_skew_data, 'Left Skewed Data')
plt.tight_layout()
plt.show()
```

# ***`Workng on Airline DEC 2008 Data using stats `***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
airline_df = pd.read_csv('/content/airline_dec_2008_50k.csv')
distance = airline_df['Distance']
distance
```

```python

mean_distance = distance.mean()
median_distance = distance.median()
skewness_distance = distance.skew()
```

```python
plt.figure(figsize = (8,6))
sns.histplot(distance , kde = True, color = 'c',bins = 30)
plt.axvline(mean_distance,color = 'r',ls ='--',label = f'Mean:{mean_distance:.2f}')
plt.axvline(median_distance,color = 'black',ls ='--',label = f'Median:{median_distance:.2f}')
```

```python
plt.title(f'Distribution of Distance\n skewness: {skewness_distance:.2f}')
plt.xlabel("Distance")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
airline_passenger = pd.read_csv('/content/airline_passenger_satisfaction.csv')
onBoard_service = airline_passenger['On-board Service']
onBoard_service
```

```python
mean_passengers = onBoard_service.mean()
median_passengers = onBoard_service.median()
skewness_passengers = onBoard_service.skew()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(onBoard_service,kde=True,color='c',bins=6)
plt.axvline(mean_passengers,color='r',ls='--',label=f'Mean:{mean_passengers:.2f}')
plt.axvline(median_passengers,color='black',ls='--',label=f'Median:{median_passengers:.2f}')
plt.title(f'Distribution of On-board Service\n skewness: {skewness_passengers:.2f}')
plt.xlabel("On-board Service")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```
