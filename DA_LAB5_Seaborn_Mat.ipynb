```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

# ***`1. Write a Python programming to display a bar chart of the popularity of programming Languages .
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7 `***

```python
pop = [22.2,17.6,8.8,8,7.7,6.7]
lang = [ 'Java', 'Python' , 'PHP' , 'JavaScript', 'c#', 'C++'  ]
plt.bar(lang,pop)
plt.minorticks_on()
plt.grid(True,alpha=0.5,ls='-',which='major', color = 'r',lw = 1)
plt.grid(True,alpha=0.1,ls='-',which='minor', color = 'g',lw = 1)
plt.xlim(-0.4)
plt.xlabel("Languages")
plt.ylabel("Popularity")
```

# ***` 9. Write a Python programming to display a bar chart of the popularity of programming Languages. Increase bottom margin.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7`***

```python
pop = [22.2,17.6,8.8,8,7.7,6.7]
lang = [ 'Java', 'Python' , 'PHP' , 'JavaScript', 'c#', 'C++'  ]
plt.bar(lang,pop)
plt.minorticks_on()
plt.subplots_adjust(bottom=0.5, top=0.9)
plt.grid(True,alpha=0.5,ls='-',which='major', color = 'r',lw = 1)
plt.grid(True,alpha=0.1,ls='-',which='minor', color = 'g',lw = 1)
plt.xlim(-0.4)
plt.xlabel("Languages")
plt.xticks(rotation=90)
plt.ylabel("Popularity")
```

# ***`10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29) `***

```python
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]
x = ['G1', 'G2', 'G3', 'G4', 'G5']

X_axis = np.arange(len(x))

plt.bar(X_axis -0.2, men, 0.4, label='Men')
plt.bar(X_axis +0.2, women, 0.4, label='Women')
plt.xticks(X_axis, x)
plt.legend()
plt.show()
```
